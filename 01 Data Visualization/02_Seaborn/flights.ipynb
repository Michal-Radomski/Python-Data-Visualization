```python
import seaborn as sns
sns.set_theme(rc={"figure.figsize": (12,8)})
flights = sns.load_dataset("flights")
# flights.head()
flights.tail()
```

```python
flights_wide = flights.pivot_table(values="passengers", index="month", columns="year", observed=False)
flights_wide.head()
```

```python
ax=sns.heatmap(flights_wide, annot=True, fmt=".0f")
ax.set_title("Flight evolution per Month/Year", size=18)
```

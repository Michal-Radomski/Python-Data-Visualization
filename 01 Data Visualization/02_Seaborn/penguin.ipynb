```python
import seaborn as sns
sns.set_theme(rc={"figure.figsize": (12,8)})
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# g=sns.PairGrid(penguins)
# g.map(sns.scatterplot)
```

```python
grid=sns.pairplot(data=penguins, hue="species", corner=True)
grid.fig.suptitle("Feature comparison in penguin species", size=24)
grid.tight_layout()
grid.fig.subplots_adjust(top=0.9)
```

```python
penguins.dropna(inplace=True)
penguins['demography']=penguins.apply(lambda row: "%s_%s" % (row["species"], row["sex"]), axis=1)
```

```python
penguins.head()
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="demography", kind="scatter", palette="hls", height=8)
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="demography", kind="kde", palette="hls", height=8, alpha=0.6, fill=True)
```

```python
joint_grid=sns.jointplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="demography",height=8)
joint_grid.plot_joint(sns.kdeplot, fill=True, alpha=0.2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt # pyplot - imperative interface

sns.set_theme(rc={"figure.figsize": (12,8)})

sns.get_dataset_names()
```

```python
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
titanic.shape
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))
sns.histplot(data=titanic, x="age", ax=axes[0,0])
sns.histplot(data=titanic, x="who", ax=axes[0,1], color="gold")
sns.histplot(data=titanic, x="class", ax=axes[1,0], color="skyblue")
sns.histplot(data=titanic, x="embark_town", ax=axes[1,1], color="teal")
fig.suptitle("Passenger demographics", size=24)
```

```python
ax=sns.kdeplot(data=titanic, x="age", hue="class", fill=True)
ax.set_title("Passenger age by class", size=24)
```

```python
ax=sns.ecdfplot(data=titanic, x="age", hue="class")
ax.set_title("Cumulative age distribution by age", size=24)
```

```python
sns.displot(data=titanic, x="who", col="class", hue="alone",multiple="stack")
```

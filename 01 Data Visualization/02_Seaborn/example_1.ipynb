```python
import pandas as pd

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Math': [85, 92, 78, 95],
    'Science': [88, 85, 92, 90]
}
df = pd.DataFrame(data)
df
```

```python
print(df.shape)  # (4, 3) - rows, columns
df.info()
df.describe()
```

```python
high_math = df[df['Math'] > 90]
print(high_math)

avg_science = df['Science'].mean()
print(f"Average Science score: {avg_science}")
```

```python
import matplotlib.pyplot as plt # pyplot - imperative interface
import numpy as np
# print(np)
import matplotlib.patches as patches
```

```python
x=np.linspace(0,10,100)
# print(x)
```

```python
plt.plot(x, x**2)
```

```python
plt.figure()
plt.plot(x,x**3)
```

```python
x = np.linspace(0,2,100)
plt.figure(figsize =(8,6) )
plt.plot(x,x, label="linear", color="#38b2a1", linewidth=2.5, linestyle="-.")
plt.plot(x,np.cos(x), label="cosine", color="#aa0074", linewidth=2.5, linestyle=":")
plt.plot(x,np.sin(x), label="sine", color="#ccbbdd", linewidth=2.5, linestyle=":")
plt.xlabel("x label", size=14)
plt.ylabel("y label", size=14)
plt.title("Pyplot interface", size=18)
plt.legend(loc="upper left")
```

```python
# Object Oriented Interface
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(0,2,100)
ax.plot(x,x, label="linear", color="#38b2a1", linewidth=2.5, linestyle="-.")
ax.plot(x,np.cos(x), label="cosine", color="#aa0074", linewidth=2.5, linestyle=":")
ax.plot(x,np.sin(x), label="sine", color="#ccbbdd", linewidth=2.5, linestyle=":")
ax.set_xlabel("x label", size=14)
ax.set_ylabel("y label", size=14)
ax.set_title("Object oriented interface", size=18)
ax.legend(loc="upper left")
```

```python
x = np.linspace(0,3,100)
plt.figure(figsize =(6,4) )
plt.plot(x, x**2)
plt.plot(x, x*2)
plt.annotate("x=0", xy=(0,0), xytext=(0,2), arrowprops=dict(facecolor="black", arrowstyle="->") )
plt.annotate("x=2", xy=(2,4), xytext=(2,6), arrowprops=dict(facecolor="black", arrowstyle="->"), bbox=dict(boxstyle="round", alpha=0.1) )
```

```python
x = np.linspace(0,3,100)
plt.plot(x, x**2, x, 2*x)
p1 = patches.Ellipse(xy=(0,0), width=0.15, height=0.6, alpha = 0.3, facecolor="#aaaaff", 
                     edgecolor="black", linewidth=2, linestyle="solid")
p2 = patches.Ellipse(xy=(2,4), width=0.15, height=0.6, alpha = 0.3, facecolor="#aaaaff", 
                     edgecolor="black", linewidth=2, linestyle="solid")
ax=plt.gca()
ax.add_patch(p1)
ax.add_patch(p2)

plt.axvline(2, color="gray", alpha=0.5, linestyle="-.")
plt.axhline(4, color="gray", alpha=0.5, linestyle="-.")
```

```python
from matplotlib.ticker import FuncFormatter
# print(FuncFormatter)

def formatter(x, pos):
    return "${:1.0f}\\pi$".format(x/np.pi)

form = FuncFormatter(formatter)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(x, np.sin(x))
plt.xlim(-7, 7)
plt.ylim(-2, 2)
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
plt.yticks([-1,0,1])
plt.grid(True)
plt.xlabel("x", size=14)
plt.ylabel("y=sin(x)", size=14)
plt.gca().xaxis.set_major_formatter(form)
```

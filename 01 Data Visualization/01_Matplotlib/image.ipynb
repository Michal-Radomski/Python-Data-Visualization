```python
import matplotlib.pyplot as plt # pyplot - imperative interface
from sklearn.datasets import load_sample_image
flower = load_sample_image("flower.jpg")
# print(flower)
```

```python
plt.figure(figsize=(12,9))
plt.imshow(flower)
plt.axis("off")
plt.show()
# plt.savefig("flower.png")
```

```python
import numpy as np
rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(flower, rgb_weights)

plt.figure(figsize=(12,9))
plt.imshow(grayscale_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
colormaps = ["PuBu_r", "gist_heat", "plasma", "coolwarm", "inferno", "RdPu"]

fig, axes = plt.subplots(3,2, figsize=(12,12))

for i,cmap in enumerate(colormaps):
    ax=axes[int(i/2), i%2]
    img = ax.imshow(grayscale_image, cmap=cmap)
    plt.axis("off")
    ax.set_title(cmap, size=16)
    cbar=fig.colorbar(img, ax=ax)
    cbar.set_ticks([])

fig.suptitle("Colormaps", size=32)
fig.tight_layout()
fig.subplots_adjust(top=0.95)
```

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
        ax.text(0.5, 0.5, "Ax_%d" % (i+1), va="center", ha="center")

fig=plt.figure()
ax=plt.subplot2grid((3,4),(0,0), 1,4)
ax=plt.subplot2grid((3,4),(1,0), 1,2)
ax=plt.subplot2grid((3,4),(1,2), 1,1)
ax=plt.subplot2grid((3,4),(1,3), 1,1)
ax=plt.subplot2grid((3,4),(2,0), 1,1)
ax=plt.subplot2grid((3,4),(2,1), 1,3)

fig.suptitle("Grid")
format_axes(fig)
```

```python
import matplotlib.gridspec as gridspec

def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
   

fig = plt.figure(figsize=(12, 8))

# use the GridSpec class
gs = gridspec.GridSpec(2, 3, figure=fig)
ax1=fig.add_subplot(gs[0,:])
ax2=fig.add_subplot(gs[1,0])
ax3=fig.add_subplot(gs[1,1])
ax4=fig.add_subplot(gs[1,2])

ax1.imshow(flower)
ax2.hist(flower[...,0].flatten(), alpha=0.3, color="r", edgecolor="r", linewidth=2)
ax3.hist(flower[...,1].flatten(), alpha=0.3, color="g", edgecolor="g", linewidth=2)
ax4.hist(flower[...,2].flatten(), alpha=0.3, color="b", edgecolor="b", linewidth=2)

fig.suptitle("Color profile", size=24)
fig.tight_layout()
fig.subplots_adjust(top=0.92)

format_axes(fig)
```

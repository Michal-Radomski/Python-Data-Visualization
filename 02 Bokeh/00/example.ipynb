```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

output_notebook()  # Enables Bokeh output in Jupyter [page:0]
```

```python
# Generate data
x = np.linspace(0, 4*np.pi, 200)
y = np.sin(x)

# Create figure
p = figure(title="Sine Wave Example", x_axis_label='x', y_axis_label='sin(x)',
           height=400, width=600, tools="pan,wheel_zoom,reset")

# Add line glyph
p.line(x, y, line_width=2, line_color="blue", legend_label="sin(x)")

show(p)  # Displays interactive plot inline [page:0][web:5]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
plt.plot(x, np.sin(x))
plt.title('Sine')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
plt.plot(x, np.cos(x))
plt.title('Cosine')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(x, np.sin(x))
ax1.set_title('Sine')
ax1.set_ylabel('Value')

ax2.plot(x, np.cos(x))
ax2.set_title('Cosine')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
x = np.arange(1, 5)

for i, ax in enumerate(axes.flat):
    if i < 3:
        ax.plot(x, x * (i+1))
        ax.set_title(f'Line {i+1}')
    else:
        ax.bar(x, [j*(i+1) for j in x])
        ax.set_title(f'Bar {i-2}')
    ax.set_xticks(x)

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.text(2, 0.8, 'Peak', fontsize=12, color='red')
plt.text(5, 0, 'Zero crossing', rotation=45)
plt.xlabel('X')
plt.ylabel('Sine')
plt.show()
```

```python
plt.plot(x, y)
plt.text(7.5, -0.5, 'Minimum', 
         ha='right', va='top',  # Right-align, top baseline
         bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5),
         fontsize=14, fontweight='bold')
plt.show()
```

```python
x_points = [1, 3, 6, 8]
y_points = [np.sin(p) for p in x_points]
plt.plot(x, y, 'b-', x_points, y_points, 'ro')

for i, (xp, yp) in enumerate(zip(x_points, y_points)):
    plt.text(xp, yp+0.1, f'P{i+1}', ha='center', fontsize=10)
plt.show()
```

```python
# Widgets
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

t = np.arange(0.0, 1.0, 0.001)
initial_freq = 3.0
line, = ax.plot(t, np.sin(2 * np.pi * initial_freq * t), lw=2)

# Create slider
ax_freq = plt.axes([0.25, 0.1, 0.65, 0.03])
slider_freq = Slider(ax_freq, 'Freq', 0.1, 10.0, valinit=initial_freq)

def update(val):
    freq = slider_freq.val
    line.set_ydata(np.sin(2 * np.pi * freq * t))
    fig.canvas.draw_idle()

slider_freq.on_changed(update)
plt.show()
```

```python
from matplotlib.widgets import Button

fig, ax = plt.subplots()
t = np.arange(0.0, 1.0, 0.001)
line, = ax.plot(t, np.sin(2 * np.pi * 3 * t), lw=2)

def color_red(event):
    line.set_color('red')
    fig.canvas.draw_idle()

def color_blue(event):
    line.set_color('blue')
    fig.canvas.draw_idle()

# Buttons
ax_red = plt.axes([0.7, 0.025, 0.1, 0.075])
ax_blue = plt.axes([0.81, 0.025, 0.1, 0.075])
button_red = Button(ax_red, 'Red')
button_blue = Button(ax_blue, 'Blue')

button_red.on_clicked(color_red)
button_blue.on_clicked(color_blue)
plt.show()
```

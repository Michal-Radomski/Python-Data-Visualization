```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 1, 10]

plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.ylabel('Values')
plt.show()
```

```python
plt.barh(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 35]
product_b = [25, 30, 25, 40]
product_c = [15, 20, 35, 30]

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x, product_a, width, label='Product A')
ax.bar(x, product_b, width, bottom=product_a, label='Product B')
ax.bar(x, product_c, width, bottom=np.array(product_a)+np.array(product_b), label='Product C')

ax.set_xlabel('Quarters')
ax.set_ylabel('Sales')
ax.set_title('Stacked Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
plt.show()
```

```python
plt.barh(categories, product_a, label='Product A')
plt.barh(categories, product_b, left=product_a, label='Product B')
plt.barh(categories, product_c, left=np.array(product_a)+np.array(product_b), label='Product C')
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = plt.cm.Set3(np.linspace(0, 1, len(labels)))

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = (30 * np.random.rand(50))**2
# print(colors)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar(label='Color scale')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

plt.boxplot([data1, data2, data3], tick_labels=['Group A', 'Group B', 'Group C'])
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=np.array([3,8,1,10])
y1=np.array([6,2,7,11])
plt.plot(y, linestyle=":", color="g", linewidth="5") # dotted
plt.plot(y1)
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.plot(x, y, marker='o', linestyle='-', label='Circle markers')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```

```python
markers = ['o', '*', '^', 's', 'D', 'x', '+', 'v']
x = range(1, 6)
for m in markers:
    plt.plot(x, [len(markers)-markers.index(m)]*5, marker=m, linestyle='', markersize=10, label=m)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
x = np.linspace(0, 10, 50)
plt.plot(x, np.sin(x), label='Sine')
plt.plot(x, np.cos(x), label='Cosine')
plt.legend(title='Trigonometric Functions')
plt.show()
```

```python
x = np.linspace(0, 10, 50)
plt.plot(x, np.sin(x), label='Sine')
plt.plot(x, np.cos(x), label='Cosine')

legend = plt.legend(title='Functions', title_fontsize=14, loc='upper right')
title = legend.get_title()
title.set_weight('bold')  # Make title bold
plt.show()
```
